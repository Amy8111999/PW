Q.1:->What is Simple Linear Regression (SLR)? Explain its purpose.
Ans:- Simple Linear Regression is a statistical method used to model the relationship between two continuous variables — one independent variable (X) and one dependent variable (Y).
The purpose of SLR is to predict the value of the dependent variable based on the value of the independent variable using a best-fit straight line.

Q.2:- What are the key assumptions of Simple Linear Regression
Ans:- There are four main assumptions are of simple linear regression which are:
1.Linearity:
The relationship between the independent variable (X) and the dependent variable (Y) must be linear.Which means the mean of the dependent variable is a straight-line function of the independent variable.
2.Independence of Errors:
The residuals (errors) are independent of each other.
3.Homoscedasticity
The variance of residuals must be constant across all levels of X.
4.Normality of Errors:
The residuals must be approximately normally distributed

Q.7:-Name and briefly describe three common evaluation metrics for regression models.
Ans:-Mean Absolute Error (MAE):
Average of the absolute differences between predicted and actual values.
MAE=1/n∑|Yi-Y^i|
Mean Squared Error (MSE):
Average of the squared differences between actual and predicted values.
MSE=1/n∑(Yi-Y^i)^2
R-squared (R²):
Measures how well the regression line explains the variability of the dependent variable.
Value ranges from 0 to 1, where 1 means perfect fit.

Q.8:=What is the purpose of the R-squared metric in regression analysis?
R-squared (R^2) represents the proportion of variance in the dependent variable that is explained by the independent variable(s) in the model.

Q.9:-Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

```python
#Ans.9
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 3, 5, 7, 11])

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Print the slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

Q.10:-How do you interpret the coefficients in a simple linear regression model?
Ans:-Intercept (Bo): The predicted value of Y when X = 0. It represents the starting point of the regression line.
Slope (Bi): The change in the dependent variable (Y) for every one-unit increase in the independent variable (X).
If Bi is positive → Y increases as X increases.
If Bi is negative → Y decreases as X increase
